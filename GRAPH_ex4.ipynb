```python
import numpy as np
from scipy.linalg import svd
m1=np.array([[2,4],[1,2]])
print(m1)
U,s,VT=svd(m1)
print("U=m*m \n ",U)
print("S=m*N \n ",s)
print("VT=n*n \n ",VT)
print(U @ np.diag(s) @ VT)
```
