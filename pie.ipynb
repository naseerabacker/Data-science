```python
import matplotlib.pyplot as plt
import numpy as np
car=['audi','benz','ford','alto']
data=[23,11,44,13]
plt.pie(data,labels=car)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
car=['audi','benz','ford','alto']
y=np.array([23,11,44,13])
myexplode=[0,0.2,0,0]
plt.pie(y,labels=car,explode=myexplode)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
car=['audi','benz','ford','alto']
y=np.array([23,11,44,13])
myexplode=[0,0.2,0,0]
plt.pie(y,labels=car,explode=myexplode,shadow=True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
car=['audi','benz','ford','alto']
y=np.array([23,11,44,13])
clr=['red','yellow','black','green']
myexplode=[0,0.2,0,0]
plt.pie(y,labels=car,explode=myexplode,shadow=True,colors=clr)
plt.show()
```
