```python
#Matrix Transformation
# 1) Rotation

from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline

a=Image.open('indexx.png')
b=np.array(a)
c=ndimage.rotate(b,45,cval=110)
d=Image.fromarray(c)
d.save('k.png')
plt.imshow(d)
```

```python
# 2) Scaling

from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline

a=Image.open('indexx.png')
b=np.array(a)
c=ndimage.zoom(b,(.5,1,1))
d=Image.fromarray(c)
d.save('image2.jpg')
plt.imshow(d)
```

```python
# 3) Shearing
#Shear x on y

from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline

a=Image.open('indexx.png')
b=np.array(a)
h,w,c=b.shape
d=np.array([[1,0,0],[.5,1,0],[0,0,1]])
e=ndimage.affine_transform(b,d)
f=Image.fromarray(e)
f.save('image3.png')
plt.imshow(f)
```

```python
# 3) Shearing
#Shear y on x

from PIL import Image
import numpy as np
from scipy import ndimage
from matplotlib import pyplot as plt
%matplotlib inline

a=Image.open('indexx.png')
b=np.array(a)
h,w,c=b.shape
d=np.array([[1,.5,0],[0,1,0],[0,0,1]])
e=ndimage.affine_transform(b,d)
f=Image.fromarray(e)
f.save('image4.jpg')
plt.imshow(f)
```
