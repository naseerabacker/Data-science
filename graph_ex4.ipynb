```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
plt.plot(x,y,marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([1,2,3,4,5])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
plt.grid()
plt.plot(x,y,marker='o',linestyle='--',color='k')
plt.show()
```
